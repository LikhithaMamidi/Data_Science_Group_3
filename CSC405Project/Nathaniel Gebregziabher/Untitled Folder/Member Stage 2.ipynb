Weekly Stats for Deaths

```python

florida_df2 = filtered_deaths[filtered_deaths['State'] == 'FL']

# Define the date columns
date_columns2 = florida_df2.columns[1:]  # Exclude the 'State' column
date_columns2 = date_columns2.tolist()
date_columns2 = [date_columns2[i:i+7] for i in range(0, len(date_columns2), 7)]

# Calculate the mean, median, and mode for each inner list
for week_columns2, week in zip(date_columns2, date_columns2):
    deaths_mean = florida_df2[week_columns2].mean(axis=1)
    deaths_median = florida_df2[week_columns2].median(axis=1)
    deaths_mode = florida_df2[week_columns2].mode(axis=1)

    # Print the results
    for mean, median, mode in zip(deaths_mean, deaths_median, deaths_mode):
        print(f"Mean, Median, Mode deaths per Week ({week[0]} to {week[-1]}):", mean, median, mode,sep=", ")
```

Plot and compare 5 other states

```python

# Filter the DataFrame to include only the rows for Florida, New York, California, Texas, and Illinois
states = ['FL', 'NY', 'CA', 'TX', 'IL','NJ']
filtered_states = filtered_deaths[filtered_deaths['State'].isin(states)]

# Define the date columns
date_columns = filtered_states.columns[1:].tolist()  # Convert to list of strings

# Calculate the mean for each state
deaths_mean = filtered_states.groupby('State')[date_columns].mean()

# Plot the results
for state, deaths in deaths_mean.iterrows():
    plt.plot(date_columns, deaths, label=state)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Mean Deaths')
plt.title('Mean Deaths per Day for FL, NY, CA, TX, NJ and IL')
plt.legend()

# Show the plot
plt.show()
```

The more populated states have more cases due to the simple fact they have major cities which are hubs for travel,making it more likley to get infected.Data very similar to US data in the sense of cases start to increase later into the year

```python
start_date = '2020-07-06'
end_date = '2020-12-20'
filtered_columns3 = [col for col in deathsdf.columns if start_date <= col <= end_date]

# Include 'State' and 'County Name' columns in the filtered columns
filtered_columns3 = ['State', 'County Name'] + filtered_columns3

# Filter the data to include only the columns in the filtered_columns2 list
filtered_cases3 = deathsdf[filtered_columns3]

# Assuming you have a DataFrame named 'deaths_df' with columns 'State', 'County Name', and the dates as columns

# Filter the data to include only the state of California
state_of_interest = 'CA'
state_data = filtered_cases3[filtered_cases3['State'] == state_of_interest]

# Calculate the total deaths for each county
county_totals = state_data.groupby('County Name').sum()

# Sort the counties based on the total deaths
sorted_counties = county_totals.sort_values(by='2020-12-20', ascending=False)

# Select the top five counties with the highest total deaths
top_five_counties = sorted_counties.head(5)
top_five_counties
```

```python
Plot top 5 counties
```

```python
date_columns = top_five_counties.columns[1:].tolist()  # Convert to list of strings

# Calculate the mean for each county
deaths_mean = top_five_counties.groupby('County Name')[date_columns].mean()

# Plot the results
for county, deaths in deaths_mean.iterrows():
    plt.plot(date_columns, deaths, label=county)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Mean Deaths')
plt.title('Mean Deaths per Day for Top 5 Infected Counties')
plt.legend()

# Show the plot
plt.show()
```

Yes it follows a very similar trend to U.S as peak rises at end of year.

```python
import pandas as pd

#Read data
enrichment_data = pd.read_csv('governors_county.csv')
covid_data = pd.read_csv('covid_confirmed_usafacts.csv')



```

# Check Covid case trend for a week

```python
utah_df = covid_data[covid_data['State'] == 'UT']
date_columns = utah_df.columns[-7:]
trends_df = utah_df[date_columns].diff(axis=1)

overall_trend = trends_df.sum().sum()

if overall_trend > 0:
    print("Cases are Increasing.")
elif overall_trend < 0:
    print("Cases are Decreasing.")
else:
    print("Cases are Stable")
```

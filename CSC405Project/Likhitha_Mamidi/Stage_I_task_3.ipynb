## Calculate COVID-19 data trends for last week of the data.



```python
import pandas as pd
import matplotlib.pyplot as plt

cases = pd.read_csv("covid_confirmed_usafacts.csv")
deaths = pd.read_csv("covid_deaths_usafacts.csv")
```

## COVID-19 data trends for last week of the data for confirmed covid cases
 * The table below illustrates the data trends for confirmed COVID-19 cases in South Carolina.
 * We specifically focus on the last week of the provided dataset.


```python
# confirmed cases
condition = cases['State'] == "SC"
result = cases.loc[condition, ['countyFIPS','County Name','State','2023-07-17','2023-07-18','2023-07-19','2023-07-20','2023-07-21','2023-07-22','2023-07-23']]
result.head()
#print(result.shape)
```

* Utilizing the above filtered data for South Carolina, we will analyze the trends for the last week.
* To conduct the analysis, we will sum the confirmed COVID-19 cases reported in South Carolina for each day during the specified week.

```python
columns_to_sum = ['2023-07-17','2023-07-18','2023-07-19','2023-07-20','2023-07-21','2023-07-22','2023-07-23']
total_summ =result[columns_to_sum].sum()

#column names and sum values
sum_dff = pd.DataFrame({'Column': total_summ.index, 'Total_Sum': total_summ.values})
sum_dff
```

```python

plt.figure(figsize=(8, 5))
plt.plot(sum_dff['Column'], sum_dff['Total_Sum'], marker='o', linestyle='-', color='b')
plt.xlabel('dates')
plt.ylabel('confirmed covid cases')
plt.title('confirmed cases for the last week of the data')
plt.grid(True)
plt.show()
```

## Analysis for last week of covid confirmed cases
* Based on the results obtained, it is evident that the confirmed cases have remained stable throughout the last week of data in South Carolina.
* The graph above indicates that no new COVID-19 cases were reported in South Carolina during the specified last week.
* In summary, we can infer that the number of COVID-19 cases neither exhibited an increase nor a decrease for the last week of data, indicating a constant level of confirmed cases.

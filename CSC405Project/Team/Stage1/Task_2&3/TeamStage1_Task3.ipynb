### Importing the COVID-19 Dataset: Utilizing Pandas to Load Data on Cases, Deaths, and Population Metrics

```python
import pandas as pd

cases = pd.read_csv("covid_confirmed_usafacts.csv")
deaths = pd.read_csv("covid_deaths_usafacts.csv")
population = pd.read_csv("covid_county_population_usafacts.csv")
```

### Covid-19 Confirmed Cases Dataframe

```python
cases.head()
```

### Covid-19 Deaths Dataframe

```python
deaths.head()
```

### Covid-19 County Population Dataframe

```python
population.head()
```

### Combine Confirmed Cases, Deaths, and Population Data to Construct an Enhanced, Comprehensive COVID-19 Dataset

```python
cases_deaths_df = pd.merge(cases, deaths, on='countyFIPS', suffixes=('_cases', '_deaths'))
covid_df = pd.merge(cases_deaths_df, population, on='countyFIPS')
covid_df.head()
```

### Export the COVID-19 Enhanced Dataset to a CSV format 

```python
covid_df.to_csv("Enhanced_COVID19_DataSet.csv", index=False)
```

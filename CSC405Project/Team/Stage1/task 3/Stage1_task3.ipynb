```python
import os
import pandas as pd

df1 = pd.read_csv("covid_confirmed_usafacts.csv")
df2 = pd.read_csv("covid_deaths_usafacts.csv")
df3 = pd.read_csv("covid_county_population_usafacts.csv")
```

team:

Merge all the three variables (cases, deaths, and population) to create a super COVID-19 datafame. Export it to a csv format.



```python
#SuperCovid19


temp=pd.merge(df1,df2,left_index=True, right_index=True, how='inner')
temper=pd.merge(temp,df3,left_index=True, right_index=True, how='inner')
temper.to_csv("super_covid.csv")

del temper['countyFIPS_y']
del temper['County Name_y']
del temper['State_y']
del temper['StateFIPS_y']
del temper['countyFIPS']
del temper['County Name']

temper.to_csv("S_covid.csv")
temper
```
